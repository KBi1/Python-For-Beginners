# **Python String Creation:**
---
---

### A string in programming is a collection of characters. These characters can be letters, numbers, symbols, or spaces. We use strings to work with words and text. In Python, you may create a string by using either a single **(' ')** or double **(" ")** quotation. If you require a long string of text, use triple quotes **(''' '''** or **""" """)** to create a string that spans multiple lines.

```python
# Strings in Python are surrounded by either single quotation marks or double quotation marks.
# 'hello' is the same as "hello".
# You can display a string literal with the print() function:

print("Hello")  # Prints the string "Hello"
print('Hello')  # Prints the string 'Hello'
```

# **Assign String to a Variable**

### Assigning a string to a variable is done with the variable name followed by an equal sign and the string:

```python
a = "Hello"  # Assigning the string "Hello" to the variable 'a'
print(a)     # Printing the value of the variable 'a'
```

```python
# String Creation

# Creating a string using single quotes and assigning it to the variable 'single_quoted'
single_quoted = 'Hello, World! '

# Creating a string using double quotes and assigning it to the variable 'double_quoted'
double_quoted = "Python is great!"

# Printing the created strings
print(single_quoted)  # Prints the string 'Hello, World! '
print(double_quoted)  # Prints the string "Python is great!"
```

# **Multiline Strings**

### Whenever you wish to define a string with more than one line, you should use triple quotation marks (''' or "") within the quotation marks:

```python
# Creating a multi-line string using three single quotes and assigning it to the variable 'multi_line_single'
multi_line_single = '''
This
is
a
multi-line
string!
'''
# Printing the multi-line string
print(multi_line_single)

# OR
print('----------- OR ----------')

# Creating a multi-line string using three double quotes and assigning it to the variable 'multi_line_double'
multi_line_double = """
This
is
a
multi-line
string!"""

# Printing the multi-line string
print(multi_line_double)
```
