# **Python String Creation**
###**Definition**:  
 ### A string in programming is a collection of characters. These characters can be **letters**, **numbers**, **symbols**, or **spaces**. Strings are used to represent and manipulate text.

### **How to Create Strings**:  
###  In Python, strings are created by enclosing characters in:
  - Single quotes (`' '`)
  - Double quotes (`" "`)
  - Triple quotes (`''' '''` or `""" """`) for multi-line strings.

### **Examples:**


```python
# Single and double quotes
# 'hello' is the same as "hello".
print("Hello")  # Output: Hello
print('Hello')  # Output: Hello
```

# **Quotes Inside Quotes**

### **Handling Quotes Inside Strings**:  
 ### To include quotes within a string:
  1. Use single quotes to enclose the string when using double quotes inside.
  2. Use double quotes to enclose the string when using single quotes inside.

 > "If you want to include quotes inside a string, ensure that the type of quotes used inside is different from those used to enclose the string."

### **Examples:**



```python
# Using single quotes with double quotes inside
sentence = 'She said, "Hello!"'
print(sentence)

# Using double quotes with single quotes inside
sentence = "He exclaimed, 'Wow!'"
print(sentence)
```


---
- **Single quotes inside double quotes**: You can use single quotes (`' '`) inside a string defined by double quotes (`" "`).
- **Double quotes inside single quotes**: You can use double quotes (`" "`) inside a string defined by single quotes (`' '`).
- **Escape characters**: Use backslashes (`\`) to include the same type of quote inside the string.

#### **Examples:**


```python
# Using double quotes inside a string enclosed by single quotes
sentence_1 = 'He said, "Hello!"'
print(sentence_1)

# Using single quotes inside a string enclosed by double quotes
sentence_2 = "She exclaimed, 'Wow!'"
print(sentence_2)

# Using escape characters to include matching quotes
sentence_3 = "He said, \"Hello!\""
print(sentence_3)
```

# **Multi-line Strings**
### **Defining Multi-line Strings:**
- **Use triple quotes** (`''' '''` or `""" """`) to create strings that span multiple lines.
- This method preserves the formatting and allows line breaks within the string.
- Both **single** (`''' '''`) and **double** (`""" """`) triple quotes can be used interchangeably.


```python
# Defining a multi-line string using triple single quotes
multi_line_single = '''
This
is
a
multi-line
string!
'''

# Printing the multi-line string
print(multi_line_single)
```

```python
# Defining a multi-line string using triple double quotes
multi_line_double = """This
is a
multi-line
string that
spans multiple
lines."""

# Printing the formatted multi-line string
print(multi_line_double)
```

## **Assigning Strings to Variables**

- **Assign Strings**:
  Strings can be stored in variables using an equal sign (`=`).

```python
a = "Hello"  # Assigning the string "Hello" to variable 'a'
print(a)
```

```python
# Assigning a string to a variable using single quotes
single_quoted = 'Hello, World!'

# Assigning a string to a variable using double quotes
double_quoted = "Python is great!"

# Printing the strings
print(single_quoted)
print(double_quoted)
```
