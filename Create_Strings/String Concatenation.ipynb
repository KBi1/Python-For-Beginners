# **String Concatenation in Python**

# **Remember that strings are immutable in Python, so each concatenation operation creates a new string object rather than modifying the original strings.**

##Strings in Python are immutable, which means they cannot be changed once they are created. Strings can be used for various operations such as  **concatenation**, **slicing**, and **formatting**. Here are some examples of common string operations that can be performed in Python.

*  ## String concatenation is the process of combining two or more strings into a single string. This operation is commonly used in programming languages to build longer strings from smaller ones.


#**Using the `**+**` operator, you can concatenate strings**

*  ## Concatenation in Python refers to the process of combining two or more strings into a single string. This operation is performed using the '+' operator.

```python
# Define two strings
string1 = "Hello"
string2 = "World"

# Concatenate the two strings
result = string1 + string2

# Print the concatenated string
print(result)  # Output: "HelloWorld"
```

## **Concatenating Strings with Space in Python**

# *Example 1:*

```python
# Define two strings
string1 = "Hello"
string2 = "World"

# Concatenate the two strings with a space in between
result = string1 + " " + string2

# Print the concatenated string
print(result)  # Output: "Hello World"
```

# *Example 2:*

```python
# In Python, two strings can be combined or joined using the "+" operator, as seen in the example code below:

# Define two strings
combined_str1 = "Hello,"
combined_str2 = "World!"

# Print the values
print(combined_str1)
print(combined_str2)

# Join the two strings
combined_string = combined_str1 + " " + combined_str2  # Adding space between the strings

# Print the concatenated string
print("Joined two different strings:", combined_string)
```

# **Multiple String Concatenations in Python**

##	- Python allows you to concatenate multiple strings in a single expression by chaining + operators.
##	- This enables you to join more than two strings at once, creating a longer string composed of smaller parts.
##  - Useful for constructing complex strings or messages dynamically.


# **You can link strings together using '+' symbols, one after another, to merge multiple strings into one.**

# *Example 1:*

```python
# Define variables for name and current date
name = "KB"
current_date = "April 28, 2024"

# Concatenate strings to create a welcome message
welcome_message = "Hello, " + name + "! Today is " + current_date + ". Welcome!"

# Print the welcome message
print(welcome_message)
```

# *Example 2:*

```python
# Combining strings using the + operator to create a full name
first_name = "Sushant"
last_name = "Rajput"
middle_name = 'Singh'

# Concatenating the first name, last name, and middle name with spaces
Full_Name = first_name + " " + last_name + " " + middle_name

# Printing the complete full name
print(Full_Name)
```

# *Example 3:*

```python
# Define variables for greeting, punctuation, and recipient
greeting = "Hello"
punctuation = ", "
recipient = "World"

# Concatenate strings to create a greeting message
result = greeting + punctuation + recipient + "!"

# Print the greeting message
print(result)  # Output: "Hello, World!"
```

# **Concatenating Variables and Strings in Python**

## In Python, you can seamlessly merge variables and strings within the same expression. This allows for the dynamic construction of strings with varying values. This makes it easier to create dynamic strings for use in various applications. For example, you can create strings with variables that will be replaced by different values at runtime. This makes it easier to create user interfaces and other interactive applications.


# **`Understanding str(age):`**

# **`The str(age) converts the integer value of age into a string, allowing it to be concatenated with other strings in the message.`**



```python
# Define variables
name = "KB"  # Assigning the value "KB" to the variable 'name'
age = 30        # Assigning the value 30 to the variable 'age'

# Concatenate strings and variables to form a message
message = "My name is " + name + " and I am " + str(age) + " years old."
# Concatenating fixed strings with the values of 'name' and 'age', converting 'age' to a string using str()

# Print the message
print(message)  # Output: "My name is KB and I am 30 years old."
# Displaying the concatenated message to the user
```

#- **F-strings in Python:**
## - An "**f-string**" (formatted string literal) is a string prefixed with 'f' or 'F'.
## - It allows embedding Python expressions directly within the string.
## - Expressions enclosed within curly braces `{}` are evaluated and replaced with their corresponding values, resulting in a dynamically constructed string.

```python
# Define variables
name = "KB"
age = 30

# Create a dynamic string using variables
message = f"Hello, my name is {name} and I am {age} years old."
# 'f' enables dynamic string formatting, allowing Python expressions within curly braces '{}'.

# Print the dynamic message
print(message)
```

# **Concatenating String Literals in Python:**

## - In Python, you can combine fixed text **(string literals)** with variables using the **`+`** operator.
## - This lets you create messages with both constant text and variable values.
## - For instance, you can mix a greeting with a person's name and age for a personalised message.
## - The **`+`** operator joins string literals and variables, making a single string.
## - This method is widely used in Python to generate dynamic text outputs.


```python
# Define a greeting
greeting = "Hello"

# Concatenate strings to form a message
result = greeting + ", " + "World" + "!"
# Concatenating the greeting with fixed strings to create the message

# Print the message
print(result)  # Output: "Hello, World!"
# Displaying the concatenated message
```

```python
# Define variables
name = "KB"
age = 30

# Concatenate string literals and variables to create a message
message = "Hello, my name is " + name + " and I am " + str(age) + " years old."

# Print the message
print(message)
```

# **Joining elements of a list into a string:**

# In Python, you can join the elements of a list into a single string using the **join()** method. This method takes a list as input and returns a string where each element of the list is concatenated together, separated by a specified delimiter.

# 1.  The **join()** method is called on a delimiter string and takes a list as its argument.
# 2.  It concatenates the elements of the list together, separated by the delimiter.
# 3.  The resulting string is returned by the **join()** method.
# 4.  This technique is useful for creating comma-separated values (CSV), constructing SQL queries, or formatting output data.




```python
# Concatenating Strings with 'join' Method
string1 = "Hello"
string2 = "World"

# Using the 'join' method to concatenate strings with a space in between
result = ' '.join([string1, string2])
# Joining the strings in the list with a space as the delimiter

# Printing the concatenated result
print(result)
```

```python
# Define a name list  of words
words = ['Sushant', 'Rajput', 'Singh']

# Using the 'join' method to concatenate the words with a space in between
joined_string = ' '.join(words)
# Joining the words in the list with a space as the delimiter

# Printing the concatenated string
print(joined_string)
```
