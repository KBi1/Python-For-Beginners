# **Modify Strings**

## There are many built-in functions in Python to manipulate strings, including changing the case (e.g., `upper()`, `lower())`, finding substrings (e.g., `find()`, `count())`, and replacing text (e.g., `replace())`, as well as other functions.


# **Upper Case**
##The **`upper()`**method in Python converts all the characters in a string to uppercase. This method does not take any parameters and returns a new string where all the alphabetic characters have been converted to uppercase

```python
# Define the original string
a = "Hello, World!"

# Convert the string to uppercase using the upper() method
print(a.upper()) # Output: "HELLO, WORLD!"
```

# **Lower Case**
## The **`lower()`** method in Python converts all the characters in a string to lowercase. Similar to the **`upper()`** method, it does not take any parameters and returns a new string where all the alphabetic characters have been converted to lowercase.

```python
# Define a string variable containing the phrase "Hello, World!"
a = "Hello, World!"

# Print the lowercase version of the string using the lower() method
print(a.lower()) # Output: "hello, world!"
```

## When we say that the **`upper()`** and **`lower()`** methods "**`do not take any parameters and return a new string,`**" we mean the following:



*  ## **Do Not Take Any Parameters:** These methods are called without any arguments inside the parentheses
* ## If you were to call them with arguments, you would get an error:





```python
a = "Hello, World!"
print(a.upper())  # Correct


# Incorrect and will cause an error
print(a.upper("arg"))
```

# **Return a New String:**
# The original string is not modified. Instead, a new string is created with all the characters converted to uppercase or lowercase. This means the original string remains unchanged, and the result of the method call is a new string.

```python
# Original string
a = "Hello, World!"

# Convert to uppercase, returns a new string
upper_case_a = a.upper()

# Convert to lowercase, returns a new string
lower_case_a = a.lower()

# Print original string
print('Original String:'+ " " + a)  # Outputs: Hello, World!

# Print the new uppercase string
print('UPPERCASE String:' + ' ' + upper_case_a)  # Outputs: HELLO, WORLD!

# Print the new lowercase string
print('lowercase String:' + ' ' + lower_case_a)  # Outputs: hello, world!
```

# **Remove Whitespace**
##Whitespace in a string is any space before **and/or** after the actual text, and often it is necessary to remove this space. The **`strip()`** method in Python removes any whitespace from the beginning and the end of a string.

```python
# Example code to demonstrate the strip() method
# String a contains whitespace at the beginning and the end.
a = " Hello, World! "
print(a.strip())  # returns "Hello, World!"
```

# **Functionality:** The strip() method removes whitespace characters from both ends of a string.
* ## The `strip()` method in Python only removes whitespace characters from the beginning and end of a string, not from within the string itself. If you want to remove whitespace from within the string as well, you can use other methods like `split()` and `join()`or regular expressions (re module).

```python
# Define a string with leading and trailing whitespace and commas
text = "              Hello, World!,    KB     "

# Use the strip() method to remove leading and trailing whitespace and commas

# Print the result
print(text.strip()) # Output: "Hello, World! , KB"
```

# **Using lstrip() to remove whitespace from the beginning:**
* ## `lstrip()` is a Python string method that removes leading whitespace (spaces, tabs, newlines) from the left side of a string.

```python
# Define a string with leading whitespace
text = "  Python Programming  !  "

# Remove leading whitespace using lstrip()
stripped_text = text.lstrip()

# Print the result
print(stripped_text)  # Output: Python Programming
```

## Using rstrip() to remove whitespace from the end:
* ## `rstrip()` is a Python string method that removes trailing whitespace (spaces, tabs, newlines) from the right side of a string.

```python
# Define a string with trailing whitespace
text = " !  Python Programming   "

# Remove trailing whitespace using rstrip()
cleaned_text = text.rstrip()

# Print the result
print(cleaned_text )  # Output: '   Python Programming'
```

# **Note:**

> ## **`strip():`** Removes whitespace from both ends of the string.
> ## **`lstrip():`** Removes whitespace only from the beginning of the string.
> ## **`rstrip():`** Removes whitespace only from the end of the string.
