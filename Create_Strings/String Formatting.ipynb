# **Python String Formatting**

# String formatting in Python means putting words and numbers together to make messages that change based on what you want to say. This lets you create messages that use different information each time. It's like filling in the blanks in a sentence with different words or numbers. This makes your messages more useful and can be adjusted for different situations.

# **" F "-strings in Python:**
## - **F-strings**, short for "**formatted string literals**",  allow embedding Python expressions directly within the string.


## -To create an f-string, simply add an **`'f'`** or **`'F'`** before the string literal, and use curly braces **`{}`** as placeholders for variables or expressions.

##- Within the curly braces **`{}`**, Python expressions are evaluated and replaced with their values, resulting in a string that dynamically adjusts based on the provided data.


```python
# Define variables
name = "KB"
age = 30

# Create a dynamic string using variables
message = f"Hello, my name is {name} and I am {age} years old."
# 'f' enables dynamic string formatting, allowing Python expressions within curly braces '{}'.

# Print the dynamic message
print(message)
```

# In this example, we're using **f-strings** to create a message that includes the person's **name**, **city**, and **age**. The variables **name**, **city**, and **age** are directly embedded within the string using curly braces **`{}`**, making the code clean and easy to read.

```python
# Define variables
name = "KB"
city = "New York"
age = 28

# Using f-strings to create a message
message = f"Hi, my name is {name}. I live in {city}, and I am {age} years old."

# Print the message
print(message)
```

# In the second example, we're using f-strings to create a message about a purchase. We have variables for the item **name** (**`item`**), its **price** (**`price`**), and the **quantity** purchased (**`quantity`**). We calculate the ***`total cost`*** by multiplying the price by the quantity. Then, we use an f-string to construct a message that includes these variables.

```python
# Define variables
item = "book"  # Name of the item
price = 19.99  # Price of each item
quantity = 3   # Quantity purchased

# Calculate total cost
total_cost = price * quantity

# Create a message using f-strings
message = f"I bought {quantity} {item}s at ${price} each. The total cost is £{total_cost:.2f}."

# Print the message
print(message)
```

# **String Formatting using `%` Operator:**

# The `%` operator in Python is used for formatting strings. It works like a fill-in-the-blank tool, where `%s` is a blank spot for a string, `%d` is a blank spot for an integer, and so on. You place your variables in a tuple after the `%` operator, and Python fills in the blanks with those values.

## **The type can be used with format codes:**

### 	1. **%d:** Use this for regular whole numbers (integers).
###2. **%f:** For numbers with decimal points (floating-point numbers).
###3.**%b:** To show numbers in binary (base 2).
### 4.**%o:** For numbers in octal (base 8).
### 5. **%x:** Used for hexadecimal (base 16) numbers.
### 6.	**%s:** When you want to display something as a simple string.
###7. 	**%e:** For numbers shown in scientific notation (exponent format).
### 8. **%r:** Placeholder for any Python object's string representation.



## ***Example 1: Basic String and Integer Formatting***

```python
name = "Alice"
age = 30

# Using the % operator to format the string
# %s is a placeholder for the 'name' variable (a string)
# %d is a placeholder for the 'age' variable (an integer)
message = "My name is %s and I am %d years old." % (name, age)

# Print the formatted message
print(message)
```

## ***Example 2: Floating-Point Numbers***

```python
price = 19.99
quantity = 3

# %f is a placeholder for floating-point numbers
# %.2f formats the floating-point number to 2 decimal places
# %d is a placeholder for the 'quantity' variable (an integer)
message = "The price per item is $%.2f and I bought %d items." % (price, quantity)

# Print the formatted message
print(message)
```

## ***Example 3: Multiple Strings***

```python
first_name = "Aqsa"
last_name = "Khan"

# %s placeholders for first_name and last_name variables (strings)
message = "Hello, my name is %s %s." % (first_name, last_name)

# Print the formatted message
print(message)
```

## ***Example 4: Hexadecimal Numbers***

```python
number = 255

# %d is a placeholder for the 'number' variable (an integer)
# %x formats the integer to a lowercase hexadecimal string
message = "The number %d in hexadecimal is %x." % (number, number)

# Print the formatted message
print(message)
```

## ***Example 5: Complex Example with Different Types***

```python
name = "KB"
age = 30
height = 5.6

# Mixed types: %s for string, %d for integer, %f for floating-point
# %.1f formats the floating-point number to 1 decimal place
message = "Name: %s, Age: %d, Height: %.1f feet" % (name, age, height)

# Print the formatted message
print(message)
```

## ***Example 6: Using %r for Any Python Object***

```python
data = [1, 2, 3]

# %r is a placeholder for the string representation of any Python object
message = "The data list is: %r" % data

# Print the formatted message
print(message)
```

# **Using Modifiers in Placeholders for String Formatting**

## A placeholder can include a modifier to format the value. A modifier is added by using a colon : followed by a formatting type.

## **Explanation:**
### - Modifiers help you control how values are formatted in a string.
### - To add a modifier, use a colon : after the placeholder.
###- Follow the colon with a formatting type.

## ***In this specific example:***

### - The placeholder is **`{price:.2f}`**.
### - **`price`**is the variable whose value (59) will be used.
### - **`.2f`** is the modifier that formats the number to have 2 decimal places.


## **Example 1: Formatting Price with Two Decimal Places Using f-String**

```python
price = 59

'''
- f-string with a placeholder
- The placeholder is {price:.2f}
- 'price' is the variable whose value will be inserted
- '.2f' is the modifier specifying that the value should be formatted as a floating-point number with 2 decimal places
'''

txt = f"The price is {price:.2f} dollars"

# Print the formatted message
print(txt)
```

## **Example 2: Calculating and Formatting Price Using f-String**

```python
# Using f-string to calculate and format the price
txt = f"The price is {10 * 59} dollars"

# Print the formatted message
print(txt)
```

## **Example 3: Formatting Percentages**

```python
# Success rate as a decimal
success_rate = 0.857

# Format the success rate as a percentage with 2 decimal places
formatted_rate = f"Success rate: {success_rate:.2%}"

# Print the formatted message
print(formatted_rate)
```

## **Example 4: Formatting Floating-Point Numbers**

```python
# Price of a product
price = 19.99

# Format the price to display with 1 decimal place
formatted_price = f"The price is ${price:.0f}"

# Print the formatted message
print(formatted_price)
```
