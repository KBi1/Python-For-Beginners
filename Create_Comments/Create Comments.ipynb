# **Creating a Comment**

### - In Python, you can create a comment by using the `#` symbol, which stands for comment. You can use comments within your code to explain, add notes, or provide documentation about what you are doing.

### - You can start a comment by using the # symbol.

### -  The Python interpreter ignores anything that follows the **`#`** symbol on the same line and considers it a comment that needs to be addressed.

### - The purpose of comments is to make your code easier to read.

```python
# Comments can explain code or provide context:

print("Hello, World!")  # This is also a single-line comment
```

# **Multiline Comments**

### - In Python, multiline comments are generally created by using triple-quotes (''' or """) instead of single-quotes.
### -  The multiline comments in Python are commonly used to explain the way functions, classes, and modules are constructed.

```python
"""
Multi-line commenting.
This can be done with single or double triple quotes.
"""
print("Hello, World!")
```

```python
'''
This is a multi-line comment or docstring.
Frequently used for documenting functions and modules, it spans multiple lines.
'''
print("Hello, World!")
```
